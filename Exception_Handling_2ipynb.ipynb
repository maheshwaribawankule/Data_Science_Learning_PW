```python
#Q.1 Explain why we have to use the Exception class while creating a Custom Exception.

class MyCustomException(Exception):
    def __init__(self, message, error_code):
        super().__init__(message)
        self.error_code = error_code

try:
    raise MyCustomException("An error occurred", 404)
except MyCustomException as e:
    print(f"{e} with error code: {e.error_code}")
```

```python
#Q.6 List down some best practices for exception handling in python.

import logging

# Set up logging
logging.basicConfig(level=logging.ERROR)

class CustomError(Exception):
    """A custom exception for demonstration purposes."""
    pass

def divide(a, b):
    try:
        return a / b
    except ZeroDivisionError as e:
        logging.error("Attempted to divide by zero.")
        raise CustomError("Division by zero is not allowed.") from e

def read_file(filename):
    try:
        with open(filename, 'r') as file:
            return file.read()
    except FileNotFoundError as e:
        logging.error(f"File not found: {filename}")
        raise CustomError(f"The file {filename} does not exist.") from e
    except IOError as e:
        logging.error(f"IO error occurred: {e}")
        raise
    finally:
        print("Completed file read attempt.")

def main():
    try:
        result = divide(10, 0)
        print(result)
    except CustomError as e:
        print(f"Custom error occurred: {e}")

    try:
        content = read_file("non_existent_file.txt")
        print(content)
    except CustomError as e:
        print(f"Custom error occurred: {e}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

if __name__ == "__main__":
    main()
```
